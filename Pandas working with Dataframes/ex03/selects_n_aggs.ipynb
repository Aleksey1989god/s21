```python
import pandas as pd
```

# 1. Load the JSON file that you created in the previous exercise into a dataframe.
- Set the CarNumber column as the index.

```python
df = pd.read_json('../data/auto.json')
df.set_index('CarNumber', inplace=True)
df.head()
```

# 2. Make the following selections:
- Display the rows where the fines are greater than 2,100.

```python
df[df['Fines'] > 2100]
```

- Display the rows where the fines are greater than 2,100 and the refund equals 2.

```python
df[(df['Fines'] > 2100) & (df['Refund'] == 2)]
```

- Display the rows where the models are from the list: [’Focus’, ’Corolla’].

```python
df[df['Model'].isin(['Focus', 'Corolla'])]
```

- Display the rows where the car number is from the list: [’Y7689C197RUS’, ’92928M178RUS’, ’7788KT197RUS’, ’H115YO163RUS’, ’X758HY197RUS’].

```python
car_numbers = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
df[df.index.isin(car_numbers)]
```

# 3. Make the aggregations with the make and model. 
- Display the median fines grouped by make.

```python
df.groupby('Make')['Fines'].median()
```

- Display the median fines grouped by make and model.

```python
df.groupby(['Make', 'Model'])['Fines'].median()
```

- Display the number of fines grouped by make and model to understand if we can trust the median values.

```python
df.groupby(['Make', 'Model'])['Fines'].count()
```

- Display the minimum and maximum fines grouped by make and model to better understand the variance.

```python
df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max']).reset_index()
```

```python
df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max'])
```

- Display the standard deviation of fines grouped by make and model to better understand the variance.

```python
df.groupby(['Make', 'Model'])['Fines'].std()
```

# 4. Make aggregations with car numbers.
- 4.1 Display the car numbers, grouped by the number of fines in descending order. We want to find those who most often violated the law.

```python
df.groupby(df.index)['Fines'].count().sort_values(ascending=False)
```

- 4.2 Select from the initial dataframe all the rows corresponding to the top-1 car number. We want to zoom in a little bit.

```python
top1_by_count_number = df.groupby(df.index)['Fines'].count().sort_values(ascending=False).index[0]
df.loc[top1_by_count_number]
```

- 4.3 Display the car numbers grouped by sum of fines in descending order. We want to find those who paid the most.

```python
df.groupby(df.index)['Fines'].sum().sort_values(ascending=False)
```

- 4.4 Select all the rows corresponding to the top-1 car number from the initial dataframe. We want to zoom in a little bit.

```python
top1_by_sum_number = df.groupby(df.index)['Fines'].sum().sort_values(ascending=False).index[0]
df.loc[top1_by_sum_number]
```

- 4.5 Display a table that answers the question, "Are there any car numbers connected to different models?"

```python
car_model_counts = df.groupby('CarNumber')['Model'].nunique()
car_model_counts[car_model_counts > 1]
```
